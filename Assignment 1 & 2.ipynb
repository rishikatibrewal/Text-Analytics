### Sentence Segmentation

Segmenting sentences using NLTK and fixing contractions , eg: don't is taken as do not instead of don\ 't

```python
def sent_segment(texts):
  text=[]
  for i in range(len(texts)):
      txt = nltk.sent_tokenize(texts[i]) 
      for j in txt:
          text.append(contractions.fix(j))
  return text
```

```python
sentences=sent_segment(reviewtext)
```

Counting the number of sentences seperated by period, exclamation mark and a question mark

```python
def count_punctuations(text):
    c=[0,0,0]
    for i in range(len(text)):
      c[0]+=len(re.findall("\.$",text[i]))
      c[1]+=len(re.findall("\?$",text[i]))
      c[2]+=len(re.findall("\!$",text[i]))
    return(c)
```

```python
c=count_punctuations(sentences)
print("Number of sentences seperated by a period is ",c[0],"\nNumber of sentences seperated by a question mark is ",c[1],"\nNumber of sentences seperated by an exclamation mark is ",c[2])
```

### Pre-processing

Using regex, separating the words, i.e, tokenizing the sentences which also removes the punctuations

```python
def tokenization(text):
    word_tokens=[]
    for i in range(len(text)):
        word_tokens.append(re.findall(r"\w+",text[i]))
    return word_tokens
```

```python
tokenized_review=tokenization(sentences)
tokenized_review[:5]
```

Removing URLs, stopword, punctuation marks and appending the lower case of the words as preprocessing steps

```python
def preprocessing(text):
    temp2=[]
    for i in range(len(text)):
        temp=[]
        for j in range(len(text[i])):

            if len(re.findall(r"(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'\".,<>?«»“”‘’]))"
,text[i][j]))!=0:
                continue
            
            elif text[i][j] in string.punctuation:
                continue
            
            elif text[i][j].lower() in stop_words:
                continue

            else:
                temp.append(text[i][j].lower())
                
        temp2.append(temp)
    return(temp2)
```

```python
preprocessed=preprocessing(tokenized_review)
```

### Frequency and rank of tokens

```python
def count_tokens(text):
    count={}
    total=0
    for i in range(len(text)):
        for j in range(len(text[i])):
            total+=1
            if text[i][j] in count.keys():
                count[text[i][j]]+=1
            else:
                count[text[i][j]]=1
    return [count,total]
```

count_tokens returns a list where 1st element is a dictionary and the latter is a positive integer. The dictionary: key is the word and value is the frequency of the word in the preprocessed corpus. The integer is the total number of tokens in the corpus.

```python
frequency=count_tokens(preprocessed)[0]
frequency
```

```python
total_count=count_tokens(preprocessed)[1]
total_count
```

Creating a dictionary of rank of tokens using the formula defined in Page 2 of https://www.ccs.neu.edu/home/ekanou/ISU535.09X2/Handouts/Review_Material/zipfslaw.pdf

```python
def rank_tokens(text):
    temp={}
    for word in frequency.keys():
        temp[word]=0.1*total_count/frequency[word]
    return temp
```

```python
rank=rank_tokens(preprocessed)
rank
```

In the preprocessing step, we removed the stopwords which had some decent frequency in the corpus. Further, we are removing the words whose frequency is less than 20 as they do not contribute significantly to the corpus when compared to the total frequency of words in the corpus

```python
new_frequency=dict(filter(lambda x:x[1]>20, frequency.items()))
```

Sorting the tokens by frequency in decreasing order

```python
sorted_frequency=dict(sorted(frequency.items(), key=lambda item: item[1],reverse=True))
sorted_frequency
```

### Pareto's Rule

Taking top 20% of the words in the dictionary sorted by frequency, then we check if together they constitute 80% of corpus i.e, cover 80% of the frequency of the tokens in the corpus.

```python
def pareto_rule(text):
    temp=list(text.values())
    c=0
    for i in range(round(len(temp)*0.2)):
        c+=temp[i]
    return(c/sum(temp))  
```

```python
pareto_rule(sorted_frequency)
```

Since the top 20% of the words cover 99% of the corpus, Pareto rule is highly applicable.
